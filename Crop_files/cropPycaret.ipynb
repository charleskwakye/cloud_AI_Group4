```python

import pandas as pd
import pycaret
crop_df = pd.read_csv('cropStats.csv',delimiter=',')
crop_df
```

we only need the location and the year so I will remove the country column and the total

```python

# Dropping the first and last columns
crop_df = crop_df.iloc[:, 1:-1]  # Selects all rows, and columns from index 1 to the second-to-last column

crop_df
```

```python
crop_df.dtypes
```

Column with categorical features is the location column, so I will set up an experiment, with the target column as the year 2017

```python
from pycaret.regression import *
reg = setup(crop_df, target = '2017')
```

```python
# # functional API
# best = compare_models()

# create linear regression model
lr = create_model('lr')

```

```python
# print(best)

# evaluate model performance
evaluate_model(lr)
```

```python
# evaluate_model(best)
# tune model hyperparameters
tuned_lr = tune_model(lr, optimize = 'r2')
```

```python
# predict_model(best)
# compare models
compare_models([lr, tuned_lr])

# # plot model performance
# plot_model(tuned_lr, plot = 'learning')
# plot_model(tuned_lr, plot = 'error')

```

As you can see the newly tuned model does better than the previous in terms of Mean absolute error, Root Mean Squared Error and Root squared error but when it came down to Mean Squared Error the previous model performed better.

<h3>Predicting number of hectares for 2018</h3>

This show the last 5 years with 2018 added, and the predictions for 2018

```python
new_data = pd.read_csv('2018_Cropdata.csv',delimiter=',')
new_data
predictions = predict_model(tuned_lr, data=new_data)
predictions = predictions.rename(columns={'prediction_label': 'predicted_2018'})
predictions = predictions[['2015', '2016', '2017', '2018', 'predicted_2018']]  # Reorder columns

last_5_columns = predictions.iloc[:,-5:]
last_5_columns
```

This shows the predicted columns where the original 2018 data was null

```python
null_2018_rows = last_5_columns[last_5_columns['2018'].isnull()]
print(null_2018_rows)
```
